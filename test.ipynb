```python
from llama_index.core import SimpleDirectoryReader, VectorStoreIndex
from llama_index.embeddings.huggingface import HuggingFaceEmbedding


documents = SimpleDirectoryReader("data").load_data()


embed_model = HuggingFaceEmbedding(model_name="BAAI/bge-small-en")  # Or any other local model


index = VectorStoreIndex.from_documents(documents, embed_model=embed_model)

print("Index successfully created using a local embedding model!")
```

```python
import ollama
from llama_index.llms.ollama import Ollama

#Ollama 
ollama_llm = Ollama(model="tinyllama", request_timeout=60)  # Use local Mistral, LLaMA, or Gemma

#query engine
query_engine = index.as_query_engine(llm=ollama_llm)


response = query_engine.query("What are the key takeaways from the data?")
print(response)
```
